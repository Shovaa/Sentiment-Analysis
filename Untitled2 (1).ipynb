```python
import tweepy
from textblob import TextBlob
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
auth=tweepy.OAuthHandler("xxxxxxxxxxxxxxxxx","xxxxxxxxxxxx")
auth.set_access_token("xxxxxxxxxx","xxxxxxxxx")
api=tweepy.API(auth,wait_on_rate_limit=True)
```

```python
posts=api.user_timeline(screen_name="*******", count=100, lang="en",tweet_mode="extended")
print("Show the 5 recent tweets:\n")
i=1
for tweet in posts[0:5]:
  print(str(i) + ') '+tweet.full_text + '\n')
  i=i+1
```

```python
#create a dataframe with a cloum called tweets
df=pd.DataFrame([tweet.full_text for tweet in posts], columns=['Tweets'])

df.head()
```

```python
def cleanTxt(text):
  text=re.sub(r'@[A-Za-z0-9]+', '',text)#remove add mention
  text=re.sub(r'#','',text)
  text=re.sub(r'RT[\s]+','',text)
  text=re.sub(r'https:\\S+', '',text)
  return text

df['Tweets']=df['Tweets'].apply(cleanTxt)

df
```

```python
#create a function to get subjectiviety
def getSubjectivity(text):
  return TextBlob(text).sentiment.subjectivity

def getPolarity(text):
  return TextBlob(text).sentiment.polarity 

df['Subjectivity']=df['Tweets'].apply(getSubjectivity)
df['Polarity']=df['Tweets'].apply(getPolarity)  


#show
df
```

```python
#wordcloud plot word cloud
allwords=' '.join([twts for twts in df['Tweets']])
wordCloud = WordCloud(width=500,height=300,random_state=21, max_font_size=120).generate(allwords)
plt.imshow(wordCloud, interpolation="bilinear")
plt.axis('off')
plt.show() 
```

```python
def getAnalysis(score):
  if score < 0:
    return 'Negative'
  elif score==0:
    return 'Neutral'
  else:
    return 'Positive'  

df['Analysis']=df['Polarity'].apply(getAnalysis)
df      
```

```python
j=1
sortedDF=df.sort_values(by=['Polarity'])
for i in range(0,sortedDF.shape[0] ):
  if (sortedDF['Analysis'][i]=='Positive'):

     print(str(j) + ')' +sortedDF['Tweets'][i])
     print()
     j= j+1  
```
